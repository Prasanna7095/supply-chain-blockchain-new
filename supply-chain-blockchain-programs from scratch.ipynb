```python
from gurobipy import Model, GRB

# Sets
warehouses = ['W1', 'W2']
customers = ['C1', 'C2', 'C3']

# Parameters
fixed_cost = {'W1': 500, 'W2': 400}
capacity = {'W1': 100, 'W2': 120}
demand = {'C1': 80, 'C2': 70, 'C3': 60}
shipping_cost = {
    ('W1', 'C1'): 4, ('W1', 'C2'): 6, ('W1', 'C3'): 9,
    ('W2', 'C1'): 5, ('W2', 'C2'): 4, ('W2', 'C3'): 7
}
penalty_cost = 20  # For unmet demand

# Create model
m = Model("Enhanced_Supply_Chain")

# Decision variables
x = m.addVars(warehouses, customers, vtype=GRB.INTEGER, name="Ship")
y = m.addVars(warehouses, vtype=GRB.BINARY, name="OpenWarehouse")
z = m.addVar(vtype=GRB.BINARY, name="Send_W2_C2")  # Helper binary
unmet = m.addVars(customers, vtype=GRB.INTEGER, name="Unmet")

# Objective: Minimize total cost
m.setObjective(
    sum(shipping_cost[w, c] * x[w, c] for w in warehouses for c in customers) +
    sum(fixed_cost[w] * y[w] for w in warehouses) +
    sum(penalty_cost * unmet[c] for c in customers),
    GRB.MINIMIZE
)

# Demand constraints (with unmet demand)
for c in customers:
    m.addConstr(sum(x[w, c] for w in warehouses) + unmet[c] == demand[c], f"Demand_{c}")

# Capacity constraints
for w in warehouses:
    m.addConstr(sum(x[w, c] for c in customers) <= capacity[w] * y[w], f"Capacity_{w}")

# Constraint A: W1 can send at most 40 units to C1
m.addConstr(x['W1', 'C1'] <= 40, "RouteLimit_W1_C1")

# Constraint B: If W2 ships to C2, it must be at least 20 units
m.addConstr(x['W2', 'C2'] >= 20 * z, "MinShipment_W2_C2")
m.addConstr(x['W2', 'C2'] <= demand['C2'] * z, "LinkZtoX_W2_C2")

# Solve
m.optimize()

# Output
print("\nShipping Plan:")
for w in warehouses:
    for c in customers:
        val = x[w, c].X
        if val > 0:
            print(f"Ship {val} units from {w} to {c}")

print("\nWarehouse Status:")
for w in warehouses:
    print(f"{w} is {'OPEN' if y[w].X > 0.5 else 'CLOSED'}")

print("\nUnmet Demand:")
for c in customers:
    if unmet[c].X > 0:
        print(f"{c}: {unmet[c].X} units unmet")

print(f"\nTotal Cost = ₹{m.objVal}")
```

```python
# Step 1: Imports
from gurobipy import Model, GRB
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Sets
warehouses = ['W1', 'W2', 'W3']
customers = ['C1', 'C2', 'C3']

# Step 3: Parameters
fixed_cost = {'W1': 500, 'W2': 400, 'W3': 300}
capacity = {'W1': 100, 'W2': 120, 'W3': 150}
demand = {'C1': 80, 'C2': 70, 'C3': 60}
shipping_cost = {
    ('W1', 'C1'): 4, ('W1', 'C2'): 6, ('W1', 'C3'): 9,
    ('W2', 'C1'): 5, ('W2', 'C2'): 4, ('W2', 'C3'): 7,
    ('W3', 'C1'): 6, ('W3', 'C2'): 5, ('W3', 'C3'): 5
}

# Step 4: Build and solve model
m = Model("SupplyChain_3Warehouses")
x = m.addVars(warehouses, customers, vtype=GRB.INTEGER, name="Ship")
y = m.addVars(warehouses, vtype=GRB.BINARY, name="OpenWarehouse")

# Objective: Minimize total cost
m.setObjective(
    sum(shipping_cost[w, c] * x[w, c] for w in warehouses for c in customers) +
    sum(fixed_cost[w] * y[w] for w in warehouses),
    GRB.MINIMIZE
)

# Constraints
for c in customers:
    m.addConstr(sum(x[w, c] for w in warehouses) == demand[c], f"Demand_{c}")
for w in warehouses:
    m.addConstr(sum(x[w, c] for c in customers) <= capacity[w] * y[w], f"Capacity_{w}")

# Solve
m.optimize()

# Step 5: Extract and display shipment table
ship_data = []
for w in warehouses:
    row = []
    for c in customers:
        row.append(x[w, c].X)
    ship_data.append(row)

df = pd.DataFrame(ship_data, index=warehouses, columns=customers)
print("\n📦 Shipment Table:")
print(df)

# Step 6: Heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(df, annot=True, cmap="YlGnBu", fmt=".0f")
plt.title("Shipping Quantities from Warehouses to Customers")
plt.xlabel("Customers")
plt.ylabel("Warehouses")
plt.tight_layout()
plt.show()

# Step 7: Cost Breakdown
shipping_total = sum(shipping_cost[w, c] * x[w, c].X for w in warehouses for c in customers)
fixed_total = sum(fixed_cost[w] * y[w].X for w in warehouses)

labels = ['Shipping Cost', 'Warehouse Fixed Cost']
sizes = [shipping_total, fixed_total]

plt.figure(figsize=(5, 5))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['#4caf50', '#ff9800'])
plt.title("Total Cost Breakdown")
plt.show()

# Step 8: Print totals
print(f"\n✅ Shipping Cost = ₹{shipping_total}")
print(f"✅ Fixed Warehouse Cost = ₹{fixed_total}")
print(f"🎯 Total Cost = ₹{m.objVal}")
```

```python
import hashlib
import datetime
import json

# Block class definition
class Block:
    def __init__(self, index, timestamp, data, previous_hash):
        self.index = index
        self.timestamp = timestamp
        self.data = data
        self.previous_hash = previous_hash
        self.hash = self.compute_hash()

    def compute_hash(self):
        block_string = json.dumps({
            "index": self.index,
            "timestamp": self.timestamp,
            "data": self.data,
            "previous_hash": self.previous_hash
        }, sort_keys=True)
        return hashlib.sha256(block_string.encode()).hexdigest()

# Blockchain class definition
class Blockchain:
    def __init__(self):
        self.chain = []
        self.create_genesis_block()

    def create_genesis_block(self):
        genesis_block = Block(0, str(datetime.datetime.now()), {"info": "Genesis Block"}, "0")
        self.chain.append(genesis_block)

    def add_block(self, data):
        previous_block = self.chain[-1]
        new_block = Block(
            index=previous_block.index + 1,
            timestamp=str(datetime.datetime.now()),
            data=data,
            previous_hash=previous_block.hash
        )
        self.chain.append(new_block)

# Create the blockchain
supply_chain = Blockchain()

# Add some shipment data
supply_chain.add_block({
    "from": "W2",
    "to": "C1",
    "quantity": 50,
    "cost": 250
})

supply_chain.add_block({
    "from": "W3",
    "to": "C3",
    "quantity": 60,
    "cost": 300
})

# Print the blocks
for block in supply_chain.chain:
    print("📦 Block:", block.index)
    print("Timestamp:", block.timestamp)
    print("Data:", block.data)
    print("Hash:", block.hash)
    print("Previous Hash:", block.previous_hash)
    print("-" * 50)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# PARAMETERS (based on paper)
# -----------------------------
a = 200      # Base demand
b = 2        # Price sensitivity
theta = 10   # Trust from blockchain
gamma = 5    # Privacy concern (reduces trust)
c = 20       # Manufacturer cost per unit

# -----------------------------
# DEMAND FUNCTION
# -----------------------------
def demand(p, B):
    return a - b * p + (theta - gamma) * B

# -----------------------------
# RETAILER'S OPTIMAL PRICING FUNCTION
# -----------------------------
def optimal_retail_price(w, B):
    return (a + (theta - gamma) * B + b * w) / (2 * b)

# -----------------------------
# RETAILER'S PROFIT
# -----------------------------
def retailer_profit(p, w, B):
    D = demand(p, B)
    return (p - w) * D

# -----------------------------
# MANUFACTURER'S PROFIT
# -----------------------------
def manufacturer_profit(w, B):
    p = optimal_retail_price(w, B)
    D = demand(p, B)
    return (w - c) * D

# -----------------------------
# GRID SEARCH TO FIND OPTIMAL w & B
# -----------------------------
best_profit = -np.inf
best_w, best_B = None, None

w_values = np.linspace(20, 80, 100)   # Try wholesale prices
B_values = np.linspace(0, 1, 50)      # Blockchain adoption levels (0 = no blockchain, 1 = full)

profit_matrix = np.zeros((len(w_values), len(B_values)))

for i, w in enumerate(w_values):
    for j, B in enumerate(B_values):
        prof = manufacturer_profit(w, B)
        profit_matrix[i][j] = prof

        if prof > best_profit:
            best_profit = prof
            best_w = w
            best_B = B

# -----------------------------
# FINAL OPTIMAL OUTPUT
# -----------------------------
optimal_p = optimal_retail_price(best_w, best_B)
optimal_d = demand(optimal_p, best_B)

print("🔍 Optimal Strategy:")
print(f"✅ Wholesale Price (w): ₹{best_w:.2f}")
print(f"✅ Blockchain Adoption (B): {best_B:.2f}")
print(f"✅ Retail Price (p*): ₹{optimal_p:.2f}")
print(f"✅ Demand: {optimal_d:.2f} units")
print(f"✅ Manufacturer Profit: ₹{best_profit:.2f}")

# -----------------------------
# OPTIONAL: VISUALIZE PROFIT SURFACE
# -----------------------------
from mpl_toolkits.mplot3d import Axes3D

W, B = np.meshgrid(w_values, B_values, indexing='ij')
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(W, B, profit_matrix, cmap='viridis')

ax.set_xlabel("Wholesale Price (w)")
ax.set_ylabel("Blockchain Level (B)")
ax.set_zlabel("Profit")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --------------------------
# Parameters from the paper
# --------------------------
a = 200
b = 2
theta = 10
gamma = 5
c = 20

# --------------------------
# Demand Function
# --------------------------
def demand(p, B):
    return a - b*p + (theta - gamma)*B

# --------------------------
# Centralized Supply Chain
# --------------------------
def centralized_model():
    best_profit = -np.inf
    for p in np.linspace(40, 120, 200):
        for B in np.linspace(0, 1, 100):
            D = demand(p, B)
            profit = (p - c) * D
            if profit > best_profit:
                best_profit = profit
                opt_p, opt_B, opt_D = p, B, D
    return opt_p, opt_B, opt_D, best_profit

# --------------------------
# Decentralized Model (Stackelberg)
# --------------------------
def decentralized_model():
    best_profit = -np.inf
    for w in np.linspace(20, 80, 100):
        for B in np.linspace(0, 1, 100):
            p = (a + (theta - gamma)*B + b*w) / (2*b)
            D = demand(p, B)
            m_profit = (w - c) * D
            if m_profit > best_profit:
                best_profit = m_profit
                opt_w, opt_B, opt_p, opt_D = w, B, p, D
    r_profit = (opt_p - opt_w) * opt_D
    return opt_w, opt_B, opt_p, opt_D, best_profit, r_profit

# --------------------------
# Coordinated Model via Smart Contract
# --------------------------
def coordinated_contract(w_prime, centralized_profit, centralized_demand):
    # Retailer's profit at w'
    p = (a + (theta - gamma) * 1 + b * w_prime) / (2 * b)
    D = demand(p, 1)
    r_profit = (p - w_prime) * D

    # Set fixed transfer T
    T = r_profit - (centralized_profit - (w_prime - c) * D)
    
    return p, D, T, (w_prime - c)*D, centralized_profit - T

# --------------------------
# Run All Models
# --------------------------
# 1. Centralized
cp, cB, cD, cProfit = centralized_model()

# 2. Decentralized
dw, dB, dp, dD, mProfit, rProfit = decentralized_model()

# 3. Coordinated via Smart Contract
w_prime = 40  # agreed lower price
contract_p, contract_D, T, m_new_profit, r_new_profit = coordinated_contract(w_prime, cProfit, cD)

# --------------------------
# Display All Results
# --------------------------
print("🔗 CENTRALIZED OPTIMUM")
print(f"Retail Price: ₹{cp:.2f}, Blockchain: {cB:.2f}, Demand: {cD:.2f}, Total Profit: ₹{cProfit:.2f}")

print("\n🔒 DECENTRALIZED (No Contract)")
print(f"Wholesale Price: ₹{dw:.2f}, Retail Price: ₹{dp:.2f}, Blockchain: {dB:.2f}")
print(f"Manufacturer Profit: ₹{mProfit:.2f}, Retailer Profit: ₹{rProfit:.2f}")

print("\n🧾 COORDINATED via Smart Contract")
print(f"New Wholesale Price: ₹{w_prime:.2f}, Retail Price: ₹{contract_p:.2f}")
print(f"Blockchain Level: 1.0, Demand: {contract_D:.2f}")
print(f"Fixed Transfer T = ₹{T:.2f}")
print(f"Manufacturer Profit: ₹{m_new_profit:.2f}, Retailer Profit: ₹{r_new_profit:.2f}")
```
